# Let's start looking at correlation

```python
m_data.corr(method='pearson') # pearson, kendall, spearman
```

```python
m_data.corr(method='kendall')
```

```python
m_data.corr(method='spearman')
```

```python
correlation_matrix = m_data.corr()
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Budget for film')
plt.show()
```
